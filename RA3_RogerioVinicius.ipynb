Considere a função $f(x) = x^3 - 6x^2 + 8x$.

* a) Encontre as raízes da função $f(x)$.
* b) Esboce o gráfico de $f(x)$ no intervalo que contém todas as raízes reais.
* c) Calcule a **área geométrica total** delimitada pelo gráfico de $f(x)$ e o eixo $x$.


a) Dada a função $f(x) = x^3 - 6x^2 + 8x$, encontrar as raízes.

Colocar em evidência:
$$f(x) = x(x^2-6x+8)$$

Resolver a equação do segundo grau:
$$f(x) = x(x^2-6x+8)$$
$$x^2-6x+8=0$$

Utilizando Bhaskara:
$$Δ=(-6)^2-4(1)(8) = 36 - 32 = 4$$

$$x=\frac{6 \pm \sqrt{4}}{2(1)} = \frac{6 \pm 2}{2}$$
Logo:

$$x=\frac{6+2}{2}=4 $$ $$x=\frac{6-2}{2} = 2$$

Portanto as raízes são:

$x=0$, $x=2$  e  $x=4$

```python
# b)
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 6*x**2 + 8*x


x = np.linspace(-1, 5, 400)

plt.figure(figsize=(8, 6))
plt.plot(x, f(x), label=r'$f(x) = x^3 - 6x^2 + 8x$', color='blue')

raizes = [0, 2, 4]
plt.scatter(raizes, [f(r) for r in raizes], color='red', zorder=5, label='Raízes')

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função f(x) = x³ - 6x² + 8x')
plt.legend()
plt.grid(True)

plt.show()
```

No gráfico, a curva é de uma função cúbica. Observamos que ela cresce, corta o eixo X, decresce e corta novamente, e por fim sobe novamente e corta novamente o eixo X até subir para o Y infinito.

a) Para encontrar os pontos nas funções $f(x) = -x^2 + 4x + 1$ e $g(x) = x - 1$, devemos:

Primeiramente igualar as funções

$$f(x)=g(x) => -x^2 + 4x + 1 = x - 1$$

Trocamos tudo para o mesmo lado
$$-x^2 + 4x + 1 - x + 1 = 0$$
$$-x^2  + 3x + 2 = 0$$

Multiplicar por $-1$:

$$x^2- 3x -2 = 0$$

Fazendo bhaskara:

$$Δ=(-3)^2-4(1)(-2) = 9 + 8 = 17$$

$$x = \frac{3 ± \sqrt{17}}{2}$$

$$\sqrt{17}\approx 4,123$$

$$x_1 = \frac{3+4.123}{2} = \frac{7.123}{2} \approx 3.56$$

Encontrar os pares ordenados:

$$y_1 = g(3.56) = 3.56 -1 = 2.56$$

$$y_2 = g(-0.56) = -0.56-1 = -1.56$$

Logo os pontos seriam aproximadamente:

$(-0,56,-1.56)$ e $(3.56, 2.56)$

```python
# Questão 3

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -x**2 + 4*x + 1


def g(x):
    return x - 1


x1 = (3 + np.sqrt(17)) / 2  # ≈ 3.56
x2 = (3 - np.sqrt(17)) / 2  # ≈ -0.56

x = np.linspace(-2, 6, 500)
x_fill = np.linspace(x2, x1, 500)

plt.figure(figsize=(8, 6))
plt.plot(x, f(x), label=r'$f(x) = -x^2 + 4x + 1$', color='blue')
plt.plot(x, g(x), label=r'$g(x) = x - 1$', color='green')
plt.fill_between(x_fill, f(x_fill), g(x_fill), color='orange', alpha=0.3, label='Região delimitada')
plt.scatter([x2, x1], [g(x2), g(x1)], color='red', zorder=5, label='Interseções')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de f(x) e g(x) com região delimitada')
plt.legend()
plt.grid(True)

plt.show()
```
